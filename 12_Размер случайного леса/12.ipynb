```python
import csv
import numpy as np
import pandas as pd


def writer(name, a) :
    with open(('answ_'+str(name)+'.txt'), 'w') as w :
        w.write(str(a))
```

```python
train = pd.read_csv('abalone.csv.csv')

train.head()
# train['AXP'].shape
```

```python
train['Sex'] = train['Sex'].map(lambda x: 1 if x=='F' else 0)
train.head()
```

```python
y_train = train['Rings']
X_train = pd.DataFrame(train.iloc[:,:-1]) 
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold, cross_val_score

n_est = range(1,51)
list_score = []

for i in n_est:
    est = RandomForestRegressor(n_estimators=i, random_state=1)
    cv = KFold(n_splits = 5, random_state=1, shuffle = True)
    score = cross_val_score(est, X_train, y_train, cv=cv, scoring="r2").mean()
    list_score.append(score)
```

```python
list_score
```

```python
leng = {}
    
for n, score in zip(n_est,list_score):
    if score > 0.52 :
        leng[n] = score
print(min(leng.keys()))
writer(121,min(leng.keys()))
```

```python
for n, score in enumerate(list_score):
    if score > 0.52:
        print(1, str(n + 1))
        break
```
