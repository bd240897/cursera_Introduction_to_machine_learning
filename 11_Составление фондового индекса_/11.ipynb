```python
import csv
import numpy as np
import pandas as pd


def writer(name, a) :
    with open(('answ_'+str(name)+'.txt'), 'w') as w :
        w.write(str(a))
```

```python
train = pd.read_csv('close_prices.csv.csv')

train.head()
# train['AXP'].shape
```

```python
from sklearn.decomposition import PCA
est = PCA(n_components = 10)
est.fit(train.iloc[:,1:])
```

```python
est.components_ 

est.explained_variance_ratio_[:-6].sum()
writer(111, 4)
```

```python
first_comp = est.transform(train.iloc[:,1:])
first_comp_df = pd.DataFrame(first_comp)[0]
first_comp_df
```

```python
train2 = pd.read_csv('djia_index.csv.csv')
train2['^DJI']
```

```python
corr = np.corrcoef(train2['^DJI'], first_comp_df.values)
corr_round = round(corr[0][1], 2)
writer(112, corr_round)
```
